## Our previous definitions:

### Nomanclature 
* $\mathcal{F}$: Pareto front
* $\tilde{\mathcal{F}}$: Approximation to Pareto front: $\tilde{\mathcal{F}} = \{y_i\}^n_{i=1}$
* $A$: attainment front
* $r$: reference point
* $\mathcal{T}$: targets: $\mathcal{T} = \{t_i\}^{n_t}_{i=1}$
* $\hat{T}$: target ideal $\hat{T} = \{\hat{t}_j\}_{j=1}^n$  where: $$\hat{t}_j = \min_{u \in T}\{u_j\}$$
* $\check{T}$: target azimuth $\check{T} = \{\check{t}_j\}_{j=1}^n$ where: $$\check{t}_j = \max_{u \in T}\{u_j\}$$


* $H(Y, r)$: Volume dominated by $Y$ referred to $r$

Region that is dominated by a set $Z$ in objective space:

$$ 
\mbox{dom}(Z) = \{ u \,|\, \exists z \in Z ~ s.t. ~ z \prec u \}
$$

### Volume definitions

$$
V_a = H(\mbox{dom}(\mathcal{T}) \cap \mbox{dom}(\tilde{\mathcal{F}}), r)
$$

$$ 
V_b = H(\mbox{dom}(\tilde{\mathcal{F}}) \setminus \mbox{dom}(\hat{T}), r)
$$

### Reference Volume definitions

$$
V_{ar} = H(\mbox{dom}(\mathcal{T}) \cap \mbox{dom}(\mathcal{F}), r)
$$

$$
V_{br} = H(\mbox{dom}(\mathcal{F}) \setminus \mbox{dom}(\hat{T}), r)
$$
