```python
from analysis_utilities import format_figures, load_result, save_fig
from analysis_utilities import PATH_TO_PROCESSED_DSAF_RESULTS, PATH_TO_PROCESSED_PAREGO_RESULTS
```

```python
# med_result_2d_name = 'wfg6_2obj_6dim_1.json'
# med_result_2d = load_result(med_result_2d_name, PATH_TO_PROCESSED_DSAF_RESULTS)


med_result_2d_name = 'wfg6_2obj_6dim_3.json'
med_result_2d = load_result(med_result_2d_name, PATH_TO_PROCESSED_DSAF_RESULTS)
med_result_2d_parego = load_result(med_result_2d_name, PATH_TO_PROCESSED_PAREGO_RESULTS)
```

```python
def plot_result(ax, x, history, label):
    mu = np.median(history, axis=0)
    lb = np.quantile(history, 0.25, axis=0)
    ub = np.quantile(history, 0.75, axis=0)
    
    ax.fill_between(x, lb, ub, alpha=0.2)
    ax.plot(x, mu, label=label)
    ax.set_xlabel(r"$n$")
    ax.set_ylabel(r"$H_{a}$")
    
    ax.set_xlim([10, 150])

fig0 = plt.figure(figsize=[4, 2])
ax0 = fig0.gca()

plot_result(ax = ax0, x = med_result_2d.dual_hpvref_hist_x, history = med_result_2d.dual_hpv_history[0], label=r"$DSAF$")
plot_result(ax = ax0, x = med_result_2d.dual_hpvref_hist_x, history = med_result_2d.dual_hpvref_history[0], label=r"$SAF$")
plot_result(ax = ax0, x = med_result_2d_parego.dual_hpv_hist_x, history = med_result_2d_parego.dual_hpv_history[0], label=r"$ParEGO$")
ax0.legend()

save_fig(fig0, filename="median_decay_plot_2obj_with_parego")
```

```python

med_result_3d_name = "wfg6_3obj_8dim_1.json"
med_result_3d = load_result(med_result_3d_name, PATH_TO_PROCESSED_DSAF_RESULTS)
med_result_3d_parego = load_result(med_result_3d_name, PATH_TO_PROCESSED_PAREGO_RESULTS)
```

```python
fig1 = plt.figure(figsize=[4, 2])
ax1 = fig1.gca(sharex = ax0, sharey=ax0)
plot_result(ax = ax1, x = med_result_3d.dual_hpvref_hist_x, history = med_result_3d.dual_hpv_history[0], label=r"$DSAF$")
plot_result(ax = ax1, x = med_result_3d.dual_hpvref_hist_x, history = med_result_3d.dual_hpvref_history[0], label=r"$SAF$")
plot_result(ax = ax1, x = med_result_3d_parego.dual_hpv_hist_x, history = med_result_3d_parego.dual_hpv_history[0], label=r"$ParEGO$")
ax1.legend()


save_fig(fig1, filename="median_decay_plot_3obj_with_parego")
```
