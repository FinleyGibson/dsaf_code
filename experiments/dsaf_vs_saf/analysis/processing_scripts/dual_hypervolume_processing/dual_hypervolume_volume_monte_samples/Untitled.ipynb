```python
import sys
import rootpath
sys.path.append(rootpath.detect())

import numpy as np
import matplotlib.pyplot as plt
import json
import os

from testsuite.analysis_tools import get_dhv_refpoint_dict, get_target_dict, strip_problem_names
```

```python
files = sorted(os.listdir("./"))
```

```python
T_D = get_target_dict()
T_D.keys()
```

```python
prob_name = "wfg4_2obj_6dim"
```

```python
P_D = get_dhv_refpoint_dict()
```

```python
prob_n, obj_n, dim_n = strip_problem_names(prob_name)
```

```python
P_D.keys()
```

```python
prob = f'wfg{prob_n}_{obj_n}obj_0'
P = P_D[prob_name+"_1"]
with open(f"./{prob}.json", "r") as infile:
    A, B = json.load(infile)
    
A = np.asarray(A)
B = np.asarray(B)
print(A.shape)
print(B.shape)

plt.scatter(*A.T, c="C0", s=2)
try:
    plt.scatter(*B.T, c="C1")
except TypeError:
    pass

plt.plot(*P[np.argsort(P[:, 0])].T, c="k")
```
